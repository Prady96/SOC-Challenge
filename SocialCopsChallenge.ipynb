```
##Use case Implementation

#loading scikit random forest classifier library
from sklearn.ensemble import RandomForestClassifier

#loading pandas
import pandas as pd

#loading numoy
import numpy as np

#setting random seed
np.random.seed(0)
```

```
## dataFrame for the land_train csv
train = pd.read_csv("/content/land_train.csv")
## first 5 rows in the head column
train.head()
```

```
## testing data
test = pd.read_csv("/content/land_test.csv")
test.head()
```

```
print("number of observations in train data: ",len(train))
print("number of observations in test data: ",len(test))
```

```
## list of the features columns for the data
features = train.columns[:12]
print(features)
```

```
# Factorise each target
y = pd.factorize(train['target'])[0]
print(y)
```

```
## Checking if theirs difference bw test and train
train.columns.equals(test.columns)
```

```
## As the ans is 'false' lets check the difference bw them
train.columns.difference(test.columns)
```

```
## Difference bw them is target
## create a columns in test dataset
test['target'] = ""
```

```
#check the test set
test
```

```
## Actual start of the prediction here
# creating a random forest classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Training the classifier
## Fitting the data and here y is target
clf.fit(train[features], y)
```

```
#Applying the trained Classifier to the test
clf.predict(test[features])
```

```
# check predicting data for the first 10
clf.predict_proba(test[features])[0:10]
```

```
## checking names for target for predicated class
test['target'].head()
```

```
preds = pd.unique(train[['target']].values.ravel('K'))[clf.predict(test[features])]
preds[0:20]
```

```
pd.crosstab(test['target'],preds,rownames=['Actual Target'],
           colnames = ['Predicted target'])
```
